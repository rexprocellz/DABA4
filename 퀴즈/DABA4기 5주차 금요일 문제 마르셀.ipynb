## 5. 네이버 수집한 기사에서 "제목"에 "연관기사"가 들어가는 데이터를 제거 합니다.
* 아래 문장을 완성해 주세요.
* <font color="red">.str.contains 를 사용하며 조건의 반대에는 앞에 ~ 표시로 표현할 수 있습니다.</red>

## 6. 위의 과정을 함수로 만든 것입니다.
* 빈칸을 완성해 수집한 table 을 넘겨주면 데이터프레임을 반환하는 함수를 완성해 주세요.

```python
def get_url(item_code, page_no):
    url = f"https://finance.naver.com/item/news_news.nhn?code={item_code}&page={page_no}&sm=title_entity_id.basic&clusterId="
    return url
```

```python
def get_finance_news(item_code, page_no):
    url = get_url(item_code, page_no)
    table = pd.read_html(url)
    if len(table) > 0:
        df = get_one_page_news(table)
        return df
```

```python
def get_one_page_news(table):
    temp_list = []
    for news in table:
        df = pd.DataFrame(news, columns=["제목", "정보제공", "날짜"])
        temp_list.append(df)
    
    df = pd.concat(temp_list)
    df = df.dropna()
    df = df[~df["제목"].str.contains("연관기사")]
    return df
```

* 아래 코드를 실행했을 때 수집된 데이터프레임이 출력된다면 잘 만들어 주신 것입니다.

```python
page_no = 2

temp = get_finance_news(item_code, page_no)
temp
```

## 7. while 문을 사용해 10페이지까지 수집하는 문장입니다.
* 위에서 만든 함수명을 써서 아래 while문을 완성해 주세요.

```python
import time
page_no = 0
news_list = []
while True:
    page_no = page_no + 1
    news = get_finance_news(item_code, page_no)
    news_list.append(news)
    time.sleep(0.5)
    if page_no > 10:
        break;
```
